```python
def show(filename, name):
    f_out = open(filename, 'r')
    lines = f_out.readlines()
    f_out.close()
    x, y_1, y_2 = [], [], []
    for line in lines:
        x.append(int(line.split(' ')[0]))
        y_1.append(float(line.split(' ')[1]))
        y_2.append(float(line.split(' ')[2]))

    print(x, '\n', y_1,'\n', y_2)

    peaks_1 = (np.argwhere(y_1 == np.amax(y_1))).flatten().tolist()
    print(peaks_1)

    peaks_2 = (np.argwhere(y_2 == np.amax(y_2))).flatten().tolist()
    print(peaks_2)
    
                        
    s = [i for i in range(len(x))]
    plt.plot(s, y_1, label=f'{name}-recovery rate')
    plt.plot(s, y_2, label=f'{name}-Silhouette score')
    for i in peaks_1:
        plt.plot(i, y_1[i], 'bo')
        if i == peaks_1[0]:
            plt.annotate(f'{y_1[i]}', xy=(i,y_1[i]), xytext=(i,y_1[i]))

    for i in peaks_2:
        plt.plot(i, y_2[i], 'yo')
        if i == peaks_2[0]:
            plt.annotate(f'{y_2[i]}', xy=(i,y_2[i]), xytext=(i,y_2[i]))


    plt.legend()
```

```python
def get_two_finals(filename):
    f_out = open(filename, 'r')
    lines = f_out.readlines()
    f_out.close()
    x, y_1, y_2 = [], [], []
    for line in lines:
        x.append(int(line.split(' ')[0]))
        y_1.append(float(line.split(' ')[1]))
        y_2.append(float(line.split(' ')[2]))

    two_final = []
    two_final.append(y_1[-1])
    two_final.append(y_2[-1])
   
    return two_final
```
